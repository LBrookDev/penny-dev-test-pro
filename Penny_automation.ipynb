```python
import webbrowser
import cv2
import numpy as np
import pyautogui
import time
import pyperclip
from mss import mss
```

```python
def copy_text():
    pyautogui.hotkey('ctrl', 'c')
    time.sleep(0.5)  # Give some time for the clipboard to update
    copied_text = pyperclip.paste()
    
    # Clean up the copied text
    copied_text = copied_text.replace('\r', '').replace('\n', '')
    
    # Handle 'warning_amber' in the copied text
    if 'warning_amber' in copied_text:
        copied_text = copied_text[:copied_text.index('warning_amber')]
    
    return copied_text

def click_image(image_path='C:\\Install\\start.png'):
    # Load the image to be found
    template = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    w, h = template.shape[::-1]

    # Give yourself a few seconds to switch to the screen with the picture
    time.sleep(5)

    # Take a screenshot using mss
    with mss() as sct:
        monitor = sct.monitors[1]  # Capture the primary monitor
        screenshot = np.array(sct.grab(monitor))
        screen_gray = cv2.cvtColor(screenshot, cv2.COLOR_BGRA2GRAY)

    # Use OpenCV to find the template in the screenshot
    result = cv2.matchTemplate(screen_gray, template, cv2.TM_CCOEFF_NORMED)
    threshold = 0.8
    loc = np.where(result >= threshold)

    if len(loc[0]) > 0:
        # Find the match with the highest y-coordinate (lowest on the screen)
        lowest_pt = max(zip(*loc[::-1]), key=lambda pt: pt[1])
        center_x = lowest_pt[0] + w // 2
        center_y = lowest_pt[1] + h // 2

        pyautogui.click(center_x, center_y)

def click_save(image_path='C:\\Install\\save.png'):
    # Load the image to be found
    template = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    w, h = template.shape[::-1]

    # Give yourself a few seconds to switch to the screen with the picture
    time.sleep(5)

    # Take a screenshot using mss
    with mss() as sct:
        monitor = sct.monitors[1]  # Capture the primary monitor
        screenshot = np.array(sct.grab(monitor))
        screen_gray = cv2.cvtColor(screenshot, cv2.COLOR_BGRA2GRAY)

    # Use OpenCV to find the template in the screenshot
    result = cv2.matchTemplate(screen_gray, template, cv2.TM_CCOEFF_NORMED)
    threshold = 0.8
    loc = np.where(result >= threshold)

    if len(loc[0]) > 0:
        # Find the match with the highest y-coordinate (lowest on the screen)
        highest_pt = min(zip(*loc[::-1]), key=lambda pt: pt[1])
        center_x = highest_pt[0] + w // 2
        center_y = highest_pt[1] + h // 2

        pyautogui.click(center_x, center_y)

def copy_penny_output(image_path='C:\\Install\\Penny.png'):
    # Load the image to be found
    template = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    w, h = template.shape[::-1]

    # Give yourself a few seconds to switch to the screen with the picture
    time.sleep(7)

    # Take a screenshot using mss
    with mss() as sct:
        monitor = sct.monitors[1]  # Capture the primary monitor
        screenshot = np.array(sct.grab(monitor))
        screen_gray = cv2.cvtColor(screenshot, cv2.COLOR_BGRA2GRAY)

    # Use OpenCV to find the template in the screenshot
    result = cv2.matchTemplate(screen_gray, template, cv2.TM_CCOEFF_NORMED)
    threshold = 0.8
    loc = np.where(result >= threshold)

    if len(loc[0]) > 0:
        # Find the match with the highest y-coordinate (lowest on the screen)
        lowest_pt = max(zip(*loc[::-1]), key=lambda pt: pt[1])
        y_coordinate = lowest_pt[1] + 30
        
        # Move the mouse to the center of the lowest matched area and click it
        pyautogui.moveTo(310, y_coordinate)
        pyautogui.click()
        pyautogui.mouseDown()
        pyautogui.dragTo(620, 954, duration=2)
        for _ in range(5):  # Adjust the range for the amount of scrolling needed
            pyautogui.scroll(-500)  # Negative value scrolls down
            time.sleep(0.5)  # Pause to allow the scroll to take effect
        pyautogui.keyDown('shift')
        pyautogui.click(x=620, y=954)
        pyautogui.keyUp('shift')
        time.sleep(1)
        pyautogui.mouseUp()

        return copy_text()

# Function to copy text to clipboard
def paste_text(text):
    pyperclip.copy(text)
    pyautogui.hotkey('ctrl', 'v')


def Copilot_response():
    pyautogui.moveTo(1071, 978, duration=1)
    pyautogui.click()
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('enter')

    time.sleep(7)
    pyautogui.hotkey('shift', 'tab')
    pyautogui.hotkey('shift', 'tab')
    pyautogui.press('enter')
    pyautogui.press('tab')
    pyautogui.press('enter')

    pyautogui.moveTo(298, 1008, duration=1)
    pyautogui.click()
    pyautogui.click()
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('enter')

    return pyperclip.paste()
```
